```python
# VALUE FROM PHOTOTRIAC DATASHEET, in SI
I_TSM = 1
dvdt = 600
V_TM = 1.8
I_FT = 0.015
```

```python
# VALUE FROM TRIAC DATASHEET, in SI
V_GT = 1.3
I_GT = 0.02
```

```python
# VALUE FROM EXTERNAL CONDITION, AC in RMS
V_AC = 220
```

$$R_1 + R_2 = \frac{V_{AC(Peak)}}{I_{TSM}} $$
$$V_{AC(Peak)} = V_{AC}\sqrt{2} $$

```python
# Calculate Minimum Resistor Value in series with triacs
R_sum = V_AC*np.sqrt(2)/I_TSM
print("Total Resistance is",round(R_sum),"Ohm")
```

$$V_T = (R_1+R_2)I_{GT} + (\frac{R_1+R_2}{R_{GK}}+1)V_{GT}+V_{TM} $$

 $$
 R_1+R_2 = R_{GK}\frac{V_T-V_{GT}-V_{TM}}{I_{GT}R_{GK}+V_{GT}}
 $$

```python
# Please check AN5114 for derivation
# Calculate
# V_T (Ripple)
V_T = 4
R_GK = 680
R_sum_2 = R_GK*(V_T-V_GT-V_TM)/(I_GT*R_GK+V_GT)
print("Total Resistance is",round(R_sum_2),"Ohm")
```

$$
\frac{dV}{dt}=\frac{V_{AC{peak}}}{R_1C_1}
$$
